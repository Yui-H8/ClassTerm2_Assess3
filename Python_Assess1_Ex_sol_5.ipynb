1. Create a pivot table showing the distribution of traffic offenses by gender.    
(That is, create a pivot table that will have traffic offenses listed in the rows, and gender in the columns, and the table itself will have the amount of traffic offenses)

```python
df.pivot_table(index='violation',columns='driver_gender', values='stop_date', aggfunc='count')
```

2. Show the distribution of traffic offenses by gender for 2013 and later

```python
where = df['stop_year'] >= 2013
df[where].pivot_table(index='violation', 
                      columns=['stop_year', 'driver_gender'], 
                      values='stop_date', 
                      aggfunc='count')
# Specify X and Y columns and values â€‹â€‹to create a pivot table
```

3. For 2013 and later, how many men / women were arrested (is_arrested) after committing a traffic offense?   
Save the result within a variable called temp_df 

```python
where = df['stop_year'] >= 2013
temp_df = df[where].pivot_table(index='is_arrested', 
                      columns=['stop_year', 'driver_gender'], 
                      values='stop_date', 
                      aggfunc='count')
temp_df
# Counting arrests by gender since 2013
```

4. Use temp_df and calculate what percentage were men arrested (is_arrested) more than women in 2015

```python
((temp_df[(2015, 'M')] - temp_df[(2015, 'F')]) / temp_df[(2015, 'F')] * 100)[True]
# Executes a calculation formula by specifying columns based on hints
```

5. Create a pivot table that shows for 2013 and later the distribution of traffic offenses for each year and gender but only for arrested drivers (is_arrested).   
Save the result within a variable called temp_df

```python
where1 = (df['stop_year'] >= 2013)
where2 = (df['is_arrested'] == True)
temp_df = df[where1 & where2].pivot_table(index='violation', 
                      columns=['stop_year', 'driver_gender'],
                      values='stop_date',
                      aggfunc='count')
temp_df
# Executes a calculation formula by specifying columns based on hints
```
